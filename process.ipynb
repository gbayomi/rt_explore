```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/fitocracy_posts.csv")
```

```python
df
```

```python
def make_numeric(row):
    return 1 if row['op_gender'] == "W" else 0
    
df['gender'] = df.apply(make_numeric, axis=1)
df['comment_text'] = df['post_text']
```

```python
k = 80000
M = df[df['op_gender']=="M"][:k//2] 
W = df[df['op_gender']=="W"][:k//2] 
both = M.append(W, ignore_index=True)
```

```python
both
```

```python
both = both.sample(frac=1, random_state=2)
both['random_number'] = np.random.randn(len(both))
train = both[both['random_number'] <= 0.6]
test_dev = both[both['random_number'] > 0.6]
dev = test_dev[:len(test_dev)//2]
test = test_dev[len(test_dev)//2:]
```

```python
train = train[["comment_text", "gender"]]
dev = dev[["comment_text", "gender"]]
test = test[["comment_text", "gender"]]
```

```python
train.to_csv("data/train_2.csv", index=False)
dev.to_csv("data/valid_2.csv", index=False)
test.to_csv("data/test_2.csv", index=False)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("full/ted_responses.csv")
```

```python
kind = 'status'

def drop_nan(row):
    
    text = str(row['post_text']).lower()
    
    if 'photo' in text:
        return 0
    if text == 'nan':
        return 0
    if row['post_type'] != kind:
        return 0
    
    return row['post_text']
```

```python
df['post_text'] = df.apply(drop_nan, axis=1)
```
